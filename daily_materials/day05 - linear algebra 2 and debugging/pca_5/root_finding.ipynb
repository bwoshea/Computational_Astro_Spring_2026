```python
import numpy as np
import scipy.linalg as linalg
import math

def f_combined(x):
    '''
    Solution is at x=y=0
    '''
    f1 = np.exp(-(x[1]**2+x[0]**2)/4)*x[0]**2
    f2 = 4.0*x[0]**2+1.5*x[1]**2

    return np.array([f1,f2])
```

```python
n_eqtns = 2

h = 1.0e-7
n_iter = 100
tol = 1.0e-6
x = 1.0*np.ones(n_eqtns)
A = np.zeros((n_eqtns,n_eqtns))

dx = 0.1
it = 0

while math.fabs(dx) > tol and it < n_iter:
    
    b = f_combined(x)
    
    for i in range(n_eqtns):
        for j in range(n_eqtns):
            hx = x[j]*h
            x[j] += hx
            c = f_combined(x)
            A[i,j] = (c[i]-b[i])/hx
    
    b *= -1
    
    d = linalg.solve(A,b)
    
    x += d
    
    dx = (d.dot(d))**0.5
    
    print("**", A, d, dx, it)
    
    it += 1
    
    
```
