```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
WriteData = False
```

```python
times = np.arange(0.0,120.1,0.1)
times[-1]
```

```python
omega = 0.3
phase = 2.0

Asig = 4.0 * np.sin(omega*times) + np.random.normal(0.0,1.0,size=times.size)
Bsig = 1.0 * np.sin(omega*times+phase)+ np.random.normal(0.0,0.5,size=times.size)
```

```python
plt.plot(times,Asig,'r.',times,Bsig,'b.')
```

```python
if WriteData:
    print("Writing data!")
    np.savetxt("two_signals.txt",np.c_[times,Asig,Bsig],fmt="%.4f",
               header="time [s], signal A [arb], signal B [arb]")
```

```python

def template_signal(t,noise=False):
    A_amp = 5.0
    A_fwhm = 1.0
    A_cent = 2.0

    B_amp = 3.0
    B_fwhm = 1.5
    B_cent = 5.0
    
    
    x = A_amp*np.exp(-(t-A_cent)**2/(2.0*A_fwhm**2)) + B_amp*np.exp(-(t-B_cent)**2/(2.0*B_fwhm**2))
    
    if noise==True:
        noise_fwhm = 0.1*min(A_amp,B_amp)
        x += np.random.normal(0.0,noise_fwhm,size=t.size)
        
    return x
```

```python
signal_nonoise = template_signal(times,noise=False)
signal_withnoise = template_signal(times,noise=True)
```

```python
plt.plot(times,signal_nonoise,'b-',linewidth=3,alpha=0.5)
plt.plot(times,signal_withnoise,'r.',alpha=0.75)
plt.xlim(0,15)
```

```python
period=(times.size-1)//8
arraysize=times.size
print(period,arraysize)
i=0

output_nonoise = np.zeros_like(times)
output_withnoise = np.zeros_like(times)

while i < arraysize-1:
    start=i
    end=i+period
    print("*",start,end,arraysize)
    output_nonoise[start:end] = signal_nonoise[0:period]
    output_withnoise[start:end] = signal_withnoise[0:period]
    
    i += period
```

```python
plt.plot(times,output_withnoise,'r-')
plt.plot(times,output_nonoise,'b-')
```

```python
if WriteData:
    print("Writing data!")
    np.savetxt("non_sinusoidal.txt",
               np.c_[times,output_nonoise,output_withnoise],fmt="%.4f",
               header="time [s], signal w/o noise [arb], signal w/noise [arb]")
```
