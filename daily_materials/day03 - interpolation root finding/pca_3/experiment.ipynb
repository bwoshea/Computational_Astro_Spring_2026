```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,200)

def f1(x):
    # should be an easy one to determine roots
    return (x-2.5)*(x+8.7)*(x+1.2)

def f2(x):
    # somewhat harderd to determine roots
    return (0.2*np.sin(2.0*x)+0.6) + 0.2*x

def f3(x):
    # good luck!
    return (np.sin(2.0*(x-4))+0.6) + 0.5*(x-4)**2

y1 = f1(x)
y2 = f2(x)
y3 = f3(x)
```

```python
import scipy.optimize as sciopt

sciopt.root(f2,2.0,method='krylov')
```

```python
plt.plot(x, y1, 'g-', x, y2, 'r-', x, y3, 'b-')
plt.ylim(-5,5)

x0 = [-10,10]
y0 = [0,0]

plt.plot(x0,y0,'k-')
#plt.xlim(-4,-3)
```

```python
import random as rand

print(f3(3.6352284118019034))
```

```python
xstart = -20.0
xend = 20.0
npoints = 2000
x = np.linspace(xstart,xend,npoints)

def analytic_func(x):
    return np.sin(1.3*x)**2 + 1.0*(np.cos(5.0*x-0.7)-0.5) + 0.2*np.exp(0.2*x) + (0.1*x)**3

y=analytic_func(x)
plt.plot(x,y)
plt.ylim(-10,10)

rand.seed(8675309)

nsample = 100
dx = (xend-xstart)/nsample
thisx = xstart

all_x = []
all_y = []

while(thisx <= xend):
    
    thisy = analytic_func(thisx)
    
    all_x.append(thisx)
    all_y.append(thisy)
    
    thisx += 2.0*dx*rand.random()

plt.plot(all_x,all_y,'k.')
```
