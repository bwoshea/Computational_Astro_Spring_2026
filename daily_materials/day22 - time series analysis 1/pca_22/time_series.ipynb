```python
import numpy as np
import numpy.random as npr
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
counts = np.loadtxt("GX.dat")

counts = np.reshape(counts,counts.size,order='C')

times = np.arange(0.0,counts.size/128.0,1.0/128.0)



def subsampler(t,c,f=0.125):
    
    bool_array = np.zeros_like(c,dtype='bool')
        
    for i in range(t.size):
        if npr.rand() < f:
            bool_array[i] = True
            
    sub_t = t[bool_array]
    sub_c = c[bool_array]
    
    return sub_t, sub_c
```

```python
sub_t, sub_c = subsampler(times,counts, f=0.1)
```

```python
plt.plot(times,counts,'bo',sub_t,sub_c,'r.')
plt.xlim(0,2)
```

```python
def autocorr(x):
    result = np.correlate(x, x, mode='full')
    return result[result.size/2:]

corr = autocorr(times)
```
