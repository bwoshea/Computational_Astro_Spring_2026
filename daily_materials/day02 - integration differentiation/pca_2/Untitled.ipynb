```python
x = 1.0

eps = 1.0

while (not x + eps == x):
    eps = eps/2.0
    print(eps)

# machine precision is 2*eps, since that was the last value for which
# 1 + eps was not 1
print(2*eps)
```

```python
def fac(x):
    if (x == 0):
        return 1
    else:
        return x*fac(x-1)


a = fac(3)
print(a)
print(a.bit_length())
```
