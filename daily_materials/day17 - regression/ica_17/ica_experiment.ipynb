```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.optimize import curve_fit
```

```python
rad1, mag1 = np.genfromtxt("NGC4406_profile.dat",skip_header=1,unpack=True)
rad2, mag2 = np.genfromtxt("NGC4551_profile.dat",skip_header=1,unpack=True)
rad3, mag3 = np.genfromtxt("NGC4472_profile.dat",skip_header=1,unpack=True)
```

```python
plt.plot(rad1,mag1,'r-',rad2,mag2,'b-',rad3,mag3,'k-')
```

```python
def func(r,mu0,Ie,Re,n):

    mu_r = mu0 - 2.5*( math.log10(Ie) - (0.868*n-0.142)*( (r/Re)**(1.0/n) - 1.0) )
    
    return mu_r
```

```python

testrad = func(rad3,np.max(mag3),10.0,269,5.99)
```

```python
plt.plot(rad3,testrad,'rx',rad3,mag3,'bo')

```

```python
popt3, pcov3 = curve_fit(func,rad3,mag3,bounds=([20,10,20,0.5],[30,30,300,16]))
```

```python
popt3
```

```python
plt.plot(rad3,func(rad3,*popt3),'rx',rad3,mag3,'b-')
```

```python
plt.plot(rad3,(func(rad3,*popt3)-mag3)/mag3)
```

```python
popt2, pcov2 = curve_fit(func,rad2,mag2,bounds=([20,10,20,0.5],[30,30,300,16]))
plt.plot(rad2,func(rad2,*popt2),'ro-',rad2,mag2,'b.-')
```

```python
plt.plot(rad2,(func(rad2,*popt2)-mag2)/mag2)
```

```python
popt1, pcov1 = curve_fit(func,rad1,mag1,bounds=([20,10,20,0.5],[30,30,300,16]))
plt.plot(rad1,func(rad1,*popt1),'rx',rad1,mag1,'b-')
print(popt1)
```
