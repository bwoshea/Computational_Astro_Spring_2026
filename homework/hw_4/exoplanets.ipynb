```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
t,f=np.loadtxt("KIC-8554498_lightcurve.dat",unpack=True,usecols=[0,1],skiprows=1)
```

```python
plt.plot(t,f)
#plt.xlim(137.5,155)
```

```python
from astropy.stats.lombscargle import LombScargle

frequency, power = LombScargle(t, f).autopower()
```

```python
plt.plot(frequency,power)
#plt.xlim(0,1)
```
